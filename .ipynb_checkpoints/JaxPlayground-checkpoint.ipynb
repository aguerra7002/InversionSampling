```python
import jax.numpy as jnp
from jax import grad, jit, jacfwd, jacrev, vmap, random
from jax.scipy.special import logsumexp
import jax.scipy.stats as stats

from numpy.random import gamma

import numpy as np
from torch.utils import data
from torchvision.datasets import MNIST
from torchvision import transforms

import time

import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
def selu(x, alpha=1.67, lambda_=1.05):
    return lambda_ * jnp.where(x > 0, x, alpha * jnp.exp(x) - alpha)

def relu(x):
    return jnp.maximum(0, x)

# A helper function to randomly initialize weights and biases
# for a dense neural network layer
def random_layer_params(m, n, key, scale=1e-2):
    w_key, b_key = random.split(key)
    return scale * random.normal(w_key, (n, m)), scale * random.normal(b_key, (n,))

# Initialize all layers for a fully-connected neural network with sizes "sizes"
def init_network_params(sizes, key):
    keys = random.split(key, len(sizes))
    return [random_layer_params(m, n, k) for m, n, k in zip(sizes[:-1], sizes[1:], keys)]

def predict(params, image):
    # per-example predictions
    activations = image
    for w, b in params[:-1]:
        outputs = jnp.dot(w, activations) + b
        activations = relu(outputs)
  
    final_w, final_b = params[-1]
    logits = jnp.dot(final_w, activations) + final_b
    return logits - logsumexp(logits)

# Make a batched version of the `predict` function
batched_predict = vmap(predict, in_axes=(None, 0))
```

```python
layer_sizes = [784, 512, 512, 10]
step_size = 0.01
num_epochs = 15
batch_size = 128
n_targets = 10
params = init_network_params(layer_sizes, random.PRNGKey(0))
```

```python
# This works on single examples
random_flattened_image = random.normal(random.PRNGKey(1), (28 * 28,))
preds = predict(params, random_flattened_image)
jnp.argmin(jnp.exp(preds))
```

```python
random_flattened_images = random.normal(random.PRNGKey(1), (10, 28 * 28))
batched_preds = batched_predict(params, random_flattened_images)
print(batched_preds.shape)
```

```python
def one_hot(x, k, dtype=jnp.float32):
    """Create a one-hot encoding of x of size k."""
    return jnp.array(x[:, None] == jnp.arange(k), dtype)
  
def accuracy(params, images, targets):
    target_class = jnp.argmax(targets, axis=1)
    predicted_class = jnp.argmax(batched_predict(params, images), axis=1)
    return jnp.mean(predicted_class == target_class)

def loss(params, images, targets):
    preds = batched_predict(params, images)
    return -jnp.mean(preds * targets)

@jit
def update(params, x, y):
    grads = grad(loss)(params, x, y)
    return [(w - step_size * dw, b - step_size * db) for (w, b), (dw, db) in zip(params, grads)]
```

```python
def numpy_collate(batch):
    if isinstance(batch[0], np.ndarray):
        return np.stack(batch)
    elif isinstance(batch[0], (tuple,list)):
        transposed = zip(*batch)
        return [numpy_collate(samples) for samples in transposed]
    else:
        return np.array(batch)

class NumpyLoader(data.DataLoader):
    def __init__(self, dataset, batch_size=1,
                shuffle=False, sampler=None,
                batch_sampler=None, num_workers=0,
                pin_memory=False, drop_last=False,
                timeout=0, worker_init_fn=None):
        super(self.__class__, self).__init__(dataset,
            batch_size=batch_size,
            shuffle=shuffle,
            sampler=sampler,
            batch_sampler=batch_sampler,
            num_workers=num_workers,
            collate_fn=numpy_collate,
            pin_memory=pin_memory,
            drop_last=drop_last,
            timeout=timeout,
            worker_init_fn=worker_init_fn)

class FlattenAndCast(object):
    def __call__(self, pic):
        return np.ravel(np.array(pic, dtype=jnp.float32))
    
```

```python
mnist_dataset = MNIST('.', download=True, train=True, transform=FlattenAndCast())
training_generator = NumpyLoader(mnist_dataset, batch_size=batch_size, num_workers=0)

# Get the full train dataset (for checking accuracy while training)
train_images = jnp.array(mnist_dataset.train_data.numpy().reshape(len(mnist_dataset.train_data), -1), dtype=jnp.float32)
train_labels = one_hot(np.array(mnist_dataset.train_labels), n_targets)
```

```python
# Get full test dataset
mnist_dataset_test = MNIST('.', download=True, train=False, transform=FlattenAndCast())

test_images = jnp.array(mnist_dataset_test.test_data.numpy().reshape(len(mnist_dataset_test.test_data), -1), dtype=jnp.float32)
test_labels = one_hot(np.array(mnist_dataset_test.test_labels), n_targets)
```

```python
# Scale the train images and test images to be the right size
train_images = train_images / 255
test_images = test_images / 255
```

```python
print(jnp.max(train_images))
```

```python
for epoch in range(num_epochs):
    start_time = time.time()
    for x, y in training_generator:
        y = one_hot(y, n_targets)
        params = update(params, x, y)
    epoch_time = time.time() - start_time
    
    train_acc = accuracy(params, train_images, train_labels)
    test_acc = accuracy(params, test_images, test_labels)
    print("Epoch {} in {:0.2f} sec".format(epoch, epoch_time))
    print("Training set accuracy {:0.3f}".format(train_acc))
    print("Test set accuracy {:0.3f}".format(test_acc))
```

```python
def hessian(fun):
    return jacfwd(jacrev(fun))
```

```python
def plot_image_transform_along_vec(vec_idx, levels):
    imgs_transformed = [img.reshape(28,28)]
    for i in range(levels):
        t1 = img + (2 ** i) * vecs[vec_idx]
        img_transformed = jnp.minimum(jnp.maximum(t1, jnp.zeros_like(t1)), jnp.ones_like(t1)).reshape(28, 28)
        imgs_transformed.append(img_transformed)

    plot_arr = jnp.hstack(imgs_transformed)
    plt.rcParams["figure.figsize"] = (20,3)
    plt.imshow(plot_arr)
    plt.title(f"x' reached by traveling along the {vec_idx}th eigenvector of H")
    plt.show()

def plot_image_transform_noisy(samples, noise_level=0.1):
    imgs_transformed = [img.reshape(28,28)]
    for sam in range(samples):
        movement = jnp.zeros_like(img)
        # JAX NEEDS TO UPGRADE THEIR GAMMA FUNCTION
        #noise = random.gamma(random.PRNGKey(int(time.time() * 1e4)), jnp.ones_like(inds) * noise_level)
        u_dim = inds.shape[0]
        noise = jnp.sqrt(gamma(22 / u_dim, 2.2, u_dim))
        movement = movement.at[inds].set(noise)
        tmp = vecs @ movement
        print(np.linalg.norm(tmp) ** 2)
        t1 = img + tmp
        img_transformed = jnp.clip(t1, 0, 1).reshape(28, 28)
        imgs_transformed.append(img_transformed)
    plot_arr = jnp.hstack(imgs_transformed)
    plt.rcParams["figure.figsize"] = (20,3)
    plt.imshow(plot_arr)
    plt.title(f"x, along with {samples} x's created with our U-sampling procedure")
    plt.show()
```

```python
u_basis = vecs[inds]
```

```python
# HOLY SHIT I MADE SOMETHING BETTER WITH JAX
@jit
def proj_vec_u(v):
    proj = jnp.zeros(v.shape)
    for basis in u_basis:
        proj = proj + basis * (jnp.dot(v, basis)) #/ jnp.dot(basis, basis))
    return proj

proj_vec_arr_u = vmap(proj_vec_u, in_axes=(0,))

projections = proj_vec_arr_u(v_i)
# for i in range(v_i.shape[0]):
#     proj = proj_vec_u(v_i[i])
```

```python
# Want to find the angle between the projections and the original vectorss, to indicate how much "projection" was done
# HOLY SHIT ANOTHER USEFUL JIT FUNCTION. I AM A GOD.
@jit
def angle_between_vectors(v1, v2):
    return jnp.arccos(jnp.clip(jnp.dot(v1, v2) / (jnp.linalg.norm(v1) * jnp.linalg.norm(v2)), -1, 1))

arr_angle_between_vectors = vmap(angle_between_vectors, in_axes=(0, 0))
angles = arr_angle_between_vectors(v_i, projections) * 180 / jnp.pi
```

```python
_, bins, _ = plt.hist(angles, bins=60, density=True)
y1 = stats.gamma.pdf(bins, a=1128, scale=0.04) # params of our gamma distribution.
#y2 = stats.cauchy.pdf(bins, loc=9.7, scale=0.8)
plt.plot(bins, y1, 'r')
#plt.plot(bins, y2, 'y')
plt.rcParams["figure.figsize"] = (6,4)
plt.ylabel("Density")
plt.xlabel("Angle from original image")
plt.show()
```
